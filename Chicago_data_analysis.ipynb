```python
df['Case Number'].count()
```

# How many crimes resulted in arrest and how many did not

### (Also, their sum should give the total number of crimes)

```python
df_arrestFalse = df[df['Arrest']==False]
print("Total number of crimes without arrest = {}".format(df_arrestFalse['Case Number'].count()))

df_arrestTrue = df[df['Arrest']==True]
print("Total number of crimes with arrest = {}".format(df_arrestTrue['Case Number'].count()))

print("Total number of crimes = {}".format(df_arrestFalse['Case Number'].count() + df_arrestTrue['Case Number'].count()) )
```

# Which unique types of crimes occured at the HOTEL/MOTEL?

```python



L = df[df['Location Description']=="HOTEL/MOTEL"]['Primary Type']

print("Number of crimes at HOTEL/MOTEL = {}".format(L.count()) )
print("\nUnique crimes = {}".format( len(L.unique().tolist())) )


print("\nList of uniue crimes at HOTEL/MOTEL: \n{}".format(L.unique().tolist()))
```

### Changing date time format from string to date time

## By using .dt we can access the date properties
## Let's group the data by hour and plot the distribution of crimes with time

```python
df_by_hour = df_dates.groupby(df["Date"].dt.hour).count()

df_by_hour
```

# Now, as we have all of the times, we can plot them on a figure

```python
df_by_hour["Arrest"].plot(kind = "bar")
```

# Here, we see that the peak of crimes happened at midnight. Then, two other peaks at 9 A.M. and 12 P.M.
