```python
# Import pandas library
import pandas as pd
```

# There are two ways we can access the data. 

1. Download the data as .csv file and rad from the local repository.
2. Provide a link to the data set.

### let's show both ways

```python
# Read the local file (in "./../" directory), and assign it to variable "df" (pandas instance - data frame)
path = "./../"
table_name = "Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012_.csv"

df = pd.read_csv(path+table_name)

# Showing first 5 rows using command "head()"
print("Showing first 5 rows\n")

df.head(5)
```

```python
# Read the online file by the URL, and assign it to variable "df" (pandas instance - data frame)
path_to_online = "https://data.cityofchicago.org/api/views/9xs2-f89t/rows.csv?accessType=DOWNLOAD"

df = pd.read_csv(path_to_online)#, header=None)

# Showing first 5 rows using command "head()"
print("Showing first 5 rows\n")

df.head(5)
```

## Some data sets might be quite large (such like data of Chicago crimes in this project ~ 1.5GB). Thus, it might be easier to access it using the second method from above.
